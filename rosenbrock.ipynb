# Metropolis-Hastings samplings for rosenbrock distribution

```python
import numpy as np
from   numpy.random import multivariate_normal as mvn
import matplotlib.pyplot as plt
```

```python
rosen      = lambda x, y: np.exp(-((1 - x)**2 + 100*(y - x**2)**2) / 20) # the target density function up to a constant
n_iters    = 10000 # number of iterations
samples    = np.empty((n_iters, 2))
```

```python
# initialize with a uniform sample
samples[0] = np.random.uniform(low=[-5, -5], high=[5, 5], size=2)
```

Take the proposal kernel $q$ to be $q(y|x) \sim N(x, \sigma^2)$.

```python
for i in range(1, n_iters):
    curr  = samples[i-1]
    prop  = curr + mvn(np.zeros(2), np.eye(2) * 0.1) # apply q(y|x)
   
    rho = rosen(*prop) / rosen(*curr) # acceptance rate
    
    if np.random.uniform() < rho:
        curr = prop
    samples[i] = curr
```

```python
k = 10
fig, ax = plt.subplots(1)
ax.plot(samples[:k, 0], samples[:k, 1], 'x')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 10)
```
